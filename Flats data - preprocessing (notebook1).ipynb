# Price

```python
df['price']
```

## Observations:-
- I observed two distinct formats for expressing prices: 'lakh' for amounts below one crore and 'cr' for amount equal to or exceeding one crore.
- To maintain consistency,I opted to represent all values using the 'cr' symbol as the common unit.
- I applied the same approcah to inspect the distribution of price categories across all value counts.

```python
df['price'].value_counts()
```

```python
df[df['price']== 'Price on Request']
```

- I examined approximately 10 flats, and in some cases, the price was specified as "price on request."  
- Consequently, I opted to eliminate these 10 rows from the dataset. The rationale behind this decision is related to the preparation of a linear regression model where the output is dependent on the price. Without a specified price, it becomes challenging to work with the data in the context of the regression model.


```python
df = df[df['price'] != 'Price on Request']
```

```python
df.head()
```

```python
# Converting price into 'cr' and 'float' datatype

def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```

```python
df.head()
```

# Address,AgePossession, Nearby Locations, Description, Furnish Details, Features , Rating 

- For these columns,more complex operations are needed, and they fall under the category of "Feature Engineering"
- Currently I am focusing on basic cleaning.
- However, I've implemented a small feature engineering step based on the "price" and "price_per_sqft"columns.
- I've introduced a new column, "Built-up Area of flat", calculated as follows:-
      Built-up Area of flat = (price * 1,00,00,000)/price_per_sqft
- This new column is placed at index location numer "4".
- This observation involves taking the price value, multiplying it by 1 crore (1,00,00,000), dissolving the year value, and then
dividing by the price per square foot.

```python
df.insert(loc=4,column='area',value=round((df['price']*10000000)/df['price_per_sqft']))
```

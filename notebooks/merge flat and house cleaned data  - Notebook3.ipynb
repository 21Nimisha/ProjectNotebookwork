```python
import numpy as np
import pandas as pd
```

```python
flats = pd.read_csv('flats_cleaned.csv')
houses = pd.read_csv('house_cleaned.csv')
```

```python
df = pd.concat([flats,houses],ignore_index=True)
```

```python
df = df.sample(df.shape[0],ignore_index=True)
```

- The overall effect of this code is to shuffle the rows of the DataFrame randomly and reassign a new index to the shuffled DataFrame
